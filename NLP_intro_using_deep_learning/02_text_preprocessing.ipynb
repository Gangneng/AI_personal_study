## 토큰화 시 고려 사항
1. 구두점이나 특수 문자 제외 X -> 의미가 있는 경우가 있음.
2. 줄임말과 단어 내 띄어쓰기가 있는 경우 하나로 인식할 수 있는 능력 소지
3. 표준 토큰화 예제 **Penn Treebank Tokenization**
  1. 하이푼으로 구성된 단어는 하나로 유짛나다.
  2. 아포스트로피로 **접어(clitic)**가 함께하는 단어는 분리한다. ex) doesn't -> doesn / n't

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

text = "Starting a home-based restaurant may be an ideal. it doesn't have a food chain or restaurant of their own."
print('트리뱅크 워드토크나이저 :',tokenizer.tokenize(text))
```

## 한국어 토큰화

한국어 토큰화는 영어 토큰화와 같지 않다. 영어는 단어 단위로 띄어쓰기가 이뤄지기 때문에 토큰화 난이도가 쉽다. 하지만 한국어는 다음과 같은 특성으로 인해 토큰화가 어려워진다.

1. 교착어의 특성
 * 한국어는 조사(을/를/이/가/등등..)라는 것이 존재한다. 어절이 독립적으로 구성된게 아닌 조사가 같이 붙는 경우가 많아 이를 분리해줘야 한다.
 * 한국어 토큰화는 형태소의 개념을 반드시 이해해야 한다. <br>형태소(morpheme) : 뜻을 가진 가장 작은 말의 단위
     * 자립 형태소 : 자립하여 사용할 수 있는 형태소로, 그 자체로 단어가 됨. 체언(명사, 대명사, 수사), 수식언(관형사, 부사), 감탄사 등.
     * 의존 형태소 : 다른 형태소와 결합하여 사용되는 형태소. 접사, 어미, 조사, 어간.
     * ex) 에디가 책을 읽었다. -> ['에디가','책을','읽었다.']
      * 자립 형태소 : 에디, 책
      * 의존 형태소 : -가, -을, 읽-, -었, -다
 * 띄어쓰기가 잘 지켜지지 않는다. 하지 않아도 이해가 잘 되기 때문에...

## 표제어 추출(Lemmatization)

표제어는 기본 사전형 단어 정도의 의미를 가짐.
ex) am, are, is의 표제어 -> be



표제어 추출 방법 : 형태학적 파싱 진행
* 형태학(morphology) : 형태소로부터 단어를 만들어가는 학문

형태소의 종류
* 어간(stem) : 단어 의미를 담고 있는 단어 핵심 부분
* 접사(affix) : 단어에 추가적 의미를 주는 부분

ex) cats -> cat(어간) / -s(접사)

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ['policy', 'doing', 'organization', 'have', 'going', 'love', 'lives', 'fly', 'dies', 'watched', 'has', 'starting']

print('표제어 추출 전 :',words)
print('표제어 추출 후 :',[lemmatizer.lemmatize(word) for word in words])
```

## 어간 추출(Stemming)



```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()

sentence = "This was not the map we found in Billy Bones's chest, but an accurate copy, complete in all things--names and heights and soundings--with the single exception of the red crosses and the written notes."
tokenized_sentence = word_tokenize(sentence)

print('어간 추출 전 :', tokenized_sentence)
print('어간 추출 후 :',[stemmer.stem(word) for word in tokenized_sentence])
```

* ANCE → 제거

```python
words = ['formalize', 'allowance', 'electricical']

print('어간 추출 전 :',words)
print('어간 추출 후 :',[stemmer.stem(word) for word in words])
```

### 랭커스터 스태머 알고리즘(Lancaster Stemmer)

아래는 두 스태머 알고리즘의 결과를 비교할 수 있다. 사용하고자 하는 코퍼스에 스태머를 적용해보고 적합한 스태머를 판단한 후 사용하면 된다.

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer

porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()

words = ['policy', 'doing', 'organization', 'have', 'going', 'love', 'lives', 'fly', 'dies', 'watched', 'has', 'starting']
print('어간 추출 전 :', words)
print('포터 스테머의 어간 추출 후:',[porter_stemmer.stem(w) for w in words])
print('랭커스터 스테머의 어간 추출 후:',[lancaster_stemmer.stem(w) for w in words])
```

## 한국어에서의 어간 추출

한국어는 5언 9품사의 구조를 갖고 있다.

* 체언 : 명사, 대명사, 수사
* 수식언 : 관형사, 부사
* 관계언 : 조사
* 독립언 : 감탄사
* 용언 : 동사, 형용사

이 중 용언은 어간(stem)과 어미(ending)의 결합으로 구성된다.

### 활용(conjugation)

* **활용** : 용언의 어간(stem)이 어미(ending)를 가지는 일
 * 규칙 활용 : 어간이 어미를 취할 때, 어미의 모습이 일정할 경우
 * 불규칙 활용 : 어간이 어미를 취할 때, 어간이나 어미의 모습이 변하는 경우
* **어간(stem)** : 용언 사용 시 원칙적으로 모양이 변하지 않는 부분. 때때로는 바뀜. (ex) 긋다. 긋고, 그어서, 그어라.)
* **어미(ending)** : 용언 어간 뒤에 붙어 활용하며 변하는 부분. 여러 문법적 기능 수행.

#### 규칙 활용

어간이 어미를 취할 때, 어간의 모습이 일정한 경우.
<br>
ex) 잡/어간 + 다/어미, 잡/어간 + 고/어미

#### 블규칙 활용

어간이 어미를 취할 때 어간의 모습이 바뀌거나 취하는 어미가 특수한 어미인 경우
<br>
ex) 듣-/들-, 돕/도우-, 곱/고우-, 잇/이-, 올/올-, 노랗/노라-  =>  어간의 형식이 달라지는 경우<br> 
ex2) 오르 + 아/어->올라, 하+아/어->하여, 이르+아/어->이르러, 푸르+아/어->푸르러  =>  일반적 어미가 아닌 특수한 어미를 취하는 경우

# 불용어 (stopword)

조사, 접미사와 같은 의미 분석을 하는데 기여하는 바가 없는 단어들을 **불용어(stopword)**라고 한다.

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from konlpy.tag import Okt
nltk.download('stopwords')
```

불용어 확인

```python
stop_words_list = stopwords.words('english')
print('불용어 개수 :', len(stop_words_list))
print('불용어 10개 출력 :',stop_words_list[:10])
```

불용어 제거
- 영어

```python
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english')) 

word_tokens = word_tokenize(example)

result = []
for word in word_tokens: 
    if word not in stop_words: 
        result.append(word) 

print('불용어 제거 전 :',word_tokens) 
print('불용어 제거 후 :',result)
```

- 한국어
  - 토큰화 후 조사, 접속사 등 제거. 그러나 명사, 형용사들도 제거되는 경우가 생김
  - 불용어 직접 정의
  - 보편적 불용어 참고 링크 : https://www.ranks.nl/stopwords/korean

```python
okt = Okt()

example = "고기를 아무렇게나 구우려고 하면 안 돼. 고기라고 다 같은 게 아니거든. 예컨대 삼겹살을 구울 때는 중요한 게 있지."
stop_words = "를 아무렇게나 구 우려 고 안 돼 같은 게 구울 때 는"

stop_words = set(stop_words.split(' '))
word_tokens = okt.morphs(example)

result = [word for word in word_tokens if not word in stop_words]

print('불용어 제거 전 :',word_tokens) 
print('불용어 제거 후 :',result)
```

# 정규 표현식

## 정규 표현식 문법

- 특수 문자
 - `.` : 한 개의 임의의 문자 (`\n` 제외)
 - `?` : 문자가 0개 또는 1개
 - `*` : 문자가 0개 이상
 - `+` : 문자가 1개 이상
 - `^` : 뒤의 문자열로 문자열 시작
 - `$` : 앞의 문자열로 문자열 끝
 - `{숫자}` : 숫자만큼 반복
 - `{숫자1, 숫자2}` : 숫자1 이상 숫자 2만큼 반복. `?`, `*`, `+`로 대체 가능
 - `{숫자, }` : 숫자 이상만큼 반복
 - `[ ]` : 대괄호 안 문자들 중 한 개 문자와 매치. 범위 지정도 가능. ex) `[a-z]`
 - `[^문자]` : 해당 문자 제외 문자 매치
 - `l` : `AlB`와 같이 쓰임. A또는 B의 의미.

- 문자 규칙
 - `\\\` : 역 슬래쉬 문자 자체 의미
 - `\\d` : 모든 숫자 의미 (=`[0-9]`)
 - `\\D` : 숫자 제외 모든 문자 의미 (=`[^0-9]`)
 - `\\s` : 공백 의미 (=`[\t\n\r\f\v]`)
 - `\\S` : 공백 제외 문자 의미 (=`[^\t\n\r\f\v]`)
 - `\\w` : 문자 또는 숫자 의미 (=`[a-zA-Z0-9]`)
 - `\\W` : 문자 또는 숫자가 아닌 문자 (=`[^a-zA-Z0-9]`)

- 모듈 함수
  - `re.compile()` : 정규표현식 컴파일 함수.
  - `re.search()` : 문자열 전체와 정규표현식과 매치되는지 검색
  - `re.match()` : 문자열 처음이 정규표현식과 매치되는지 검색
  - `re.split()` : 정규 표현식 기준 문자열 분리하여 리스트로 리턴
  - `re.findall()` : 정규 표현식과 매치되는 모든 경우의 문자열을 찾아 리스트로 리턴
  - `re.finditer()` : 문자열에서 정규 표현식과 매치되는 모든 경우의 문자열에 대한 이터레이터 객체 리턴
  - `re.sub()` : 문자열에서 정규 표현식과 일치하는 부분에 대해 다른 문자열 대체

정규 표현식을 이용한 토큰화

```python
from nltk.tokenize import RegexpTokenizer

text = "Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop"

tokenizer1 = RegexpTokenizer("[\w]+")
tokenizer2 = RegexpTokenizer("\s+", gaps=True)

print(tokenizer1.tokenize(text))
print(tokenizer2.tokenize(text))
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
raw_text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
```

```python
sentences = sent_tokenize(raw_text)
print(sentences)
```

```python
vocab = {}
preprocessed_sentences = []
stop_words = set(stopwords.words('english'))

for sentence in sentences:
    # 단어 토큰화
    tokenized_sentence = word_tokenize(sentence)
    result = []

    for word in tokenized_sentence: 
        word = word.lower() # 모든 단어를 소문자화하여 단어의 개수를 줄인다.
        if word not in stop_words: # 단어 토큰화 된 결과에 대해서 불용어를 제거한다.
            if len(word) > 2: # 단어 길이가 2이하인 경우에 대하여 추가로 단어를 제거한다.
                result.append(word)
                if word not in vocab:
                    vocab[word] = 0 
                vocab[word] += 1
    preprocessed_sentences.append(result) 
print(preprocessed_sentences)
```

```python
print('단어 집합 :',vocab)
```

```python
vocab_sorted = sorted(vocab.items(), key = lambda x:x[1], reverse = True)
print(vocab_sorted)
```

높은 빈도수의 단어일수록 낮은 정수 부여

```python
word_to_index = {}
i = 0
for (word, frequency) in vocab_sorted :
    if frequency > 1 : # 빈도수가 작은 단어는 제외.
        i = i + 1
        word_to_index[word] = i

print(word_to_index)
```

빈도수 상위 5개의 단어만 사용한다 가정

```python
vocab_size = 5

# 인덱스가 5 초과인 단어 제거
words_frequency = [word for word, index in word_to_index.items() if index >= vocab_size + 1]

# 해당 단어에 대한 인덱스 정보를 삭제
for w in words_frequency:
    del word_to_index[w]
print(word_to_index)
```

### OOV (Out-Of-Vocabulary)

단어 집합에 없는 단어

```python
word_to_index['OOV'] = len(word_to_index) + 1
print(word_to_index)
```

모든 단어들을 맵핑되는 정수로 인코딩

```python
encoded_sentences = []
for sentence in preprocessed_sentences:
    encoded_sentence = []
    for word in sentence:
        try:
            # 단어 집합에 있는 단어라면 해당 단어의 정수를 리턴.
            encoded_sentence.append(word_to_index[word])
        except KeyError:
            # 만약 단어 집합에 없는 단어라면 'OOV'의 정수를 리턴.
            encoded_sentence.append(word_to_index['OOV'])
    encoded_sentences.append(encoded_sentence)
print(encoded_sentences)
```

# 원-핫 인코딩(One-Hot Encoding)


## 단어 집합(vocabulary)

**서로 다른 단어들의 집합**
<br>여기서 다른 단어는 book과 books와 같이 단어 변형 형태도 다른 단어로 간주.
<br>모든 단어 중복을 허용하지 않고 모아놓은 것.

## 원-핫 인코딩이란?

표현하고 싶은 단어의 인덱스에 1을 부여. 나머지엔 0 부여해서 표현된 벡터.

```python
from konlpy.tag import Okt  

okt = Okt()  
tokens = okt.morphs("나는 자연어 처리를 배운다")  
print(tokens)
```

```python
word_to_index = {word : index for index, word in enumerate(tokens)}
print('단어 집합 :',word_to_index)
```

```python
def one_hot_encoding(word, word_to_index):
  one_hot_vector = [0]*(len(word_to_index))
  index = word_to_index[word]
  one_hot_vector[index] = 1
  return one_hot_vector
```

```python
one_hot_encoding("자연어", word_to_index)
```

### SOYNLP의 응집 확률(cohesion probability)

내부 문자열이 얼마나 응집해 자주 등장하는지를 판단하는 척도.

$$cohesion(n) = \big(\prod_{i=1}^{n-1}P(c_{1:i+1}|c_{1:i})\big)^{\frac{1}{n-1}}$$

'반포한강공원'의 응집 확률을 구해보자.
